```python
from pathlib import Path

import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
DATA_FILE_PATH = Path.cwd() / "data"
MINILM_RESULTS = DATA_FILE_PATH / "ai_stems_with_sims-all-MiniLM-L6-v2.csv"

df = pl.read_csv(MINILM_RESULTS)
plt.figure(figsize=(16, 8))

sns.violinplot(data=df, x="facet", y="cosine_similarity", inner="quartile")
plt.axhline(0.75, color="red", linestyle="--", label="Drift Threshold (0.75)")
plt.xticks(rotation=90)
plt.title("Cosine Similarities Between Anchor and Generated Stems")
plt.ylabel("Cosine Similarity")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
DATA_FILE_PATH = Path.cwd() / "data"
MINILM_RESULTS = DATA_FILE_PATH / "ai_stems_with_sims-all-MiniLM-L6-v2.csv"

df = pl.read_csv(MINILM_RESULTS)
plt.figure(figsize=(8, 5))
sns.histplot(df["cosine_similarity"], bins=30, kde=True)
plt.axvline(0.75, color="red", linestyle="--")
plt.title("Distribution of Cosine Similarities (All Facets)")
plt.xlabel("Cosine Similarity")
plt.ylabel("Stem Count")
plt.show()
```
